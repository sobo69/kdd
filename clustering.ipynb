```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Species'] = pd.Categorical.from_codes(data.target, data.target_names)

X = df[['sepal length (cm)', 'sepal width (cm)']].values

K = 3
kmeans_model = KMeans(n_clusters=K).fit(X)

centers = np.array(kmeans_model.cluster_centers_)
labels = kmeans_model.labels_


plt.plot()
plt.title('k means centroids')

colors = ['b', 'g', 'c']
markers = ['o', 'v', 's']


species_names = df['Species'].unique()

labels_ = species_names

for i, l in enumerate(labels):
    plt.plot(X[i, 0], X[i, 1], color=colors[l], marker=markers[l], ls='None', label=labels_[l] if i == labels.tolist().index(l) else "")

plt.scatter(centers[:, 0], centers[:, 1], marker="x", color='r', label="Cluster Centers")
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(loc='best')
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Data points
x1 = np.array([3,3,1,1,4,2,5,7,6,2,3,1,1,3,4,5,2])
x2 = np.array([5,4,6,6,5,8,6,7,6,7,1,2,1,2,3,2,3])

# Combine into a single array
X = np.column_stack((x1, x2))

# Apply KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
labels  = kmeans.labels_
centers = kmeans.cluster_centers_

# Plot each cluster with its own color/marker and label
colors  = ['b', 'g', 'c']
markers = ['o', 'v', 's']

for cluster in range(3):
    pts = X[labels == cluster]
    plt.scatter(
        pts[:,0], pts[:,1],
        c=colors[cluster],
        marker=markers[cluster],
        label=f'Class {cluster+1}'
    )

# Plot the cluster centers
plt.scatter(
    centers[:,0], centers[:,1],
    c='r', marker='x', s=100,
    label='Centroids'
)

plt.legend()
plt.title('KMeans Clustering')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.show()
```
